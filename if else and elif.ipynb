```python
#if statements
num=5

if num == 0:
    print('1 condition is true')

if num != 0 :
    print('2 condition is true')

if num < 0:
    print('3 condition is true')

if num > 0:
    print('4 condition is true')

if num <=0:
    print('5 condition is true')

if num >= 0:
    print('6 condition is true')
```

```python
#id-else statements
if num >6:
    print('condition is true')
else:
    print('condition is false')
```

```python
#elif statemets for miltiple conditions
if num ==0:
    print('1 stament is true')
elif num != 0:
    print('2 statement is true')
elif num< 0:
    print('3 statement is true')
elif num > 0:
    print('4 statement is true')
else:
    print('no match found')
```

```python
# (and, or, not) comparison opreator
user = 'admin'
logged_in= True

if user=='admin' and logged_in:
    print('admin page')
else:
    print('wrong credentials')
```

```python
logged_in= False

if user=='admin' or logged_in:
    print('admin page')
else:
    print('wrong credentials')
```

```python
if not logged_in:
    print('wrong cerdentials')
else:
    print('welcome')
```
